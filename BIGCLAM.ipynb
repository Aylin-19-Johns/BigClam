```python
import networkx as nx
import community.community_louvain as cl
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
G = nx.karate_club_graph()
print(f"Graph G created with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")
```

```python

partition = cl.best_partition(G)
modularity_score = cl.modularity(partition, G)
print(f"Modularity: {modularity_score:.4f}")
```

```python
print("Node memberships (community assignments):")
for node, comm_id in partition.items():
    print(f"Node {node}: Community {comm_id}")
```

```python
num_communities = len(set(partition.values()))
print(f"Number of communities found: {num_communities}")
```

```python
import matplotlib.pyplot as plt


unique_communities = set(partition.values())

cmap = plt.get_cmap('viridis', max(unique_communities) + 1)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
node_colors = [cmap(partition[node]) for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=700, alpha=0.9)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')
plt.title("Karate Club Graph with Louvain Communities")
plt.axis('off')
plt.show()
```
